```python
tr_score = list()
te_score = list()
```

```python
tree_depths = [i for i in range(1,27)]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
for i in tree_depths:

    decition_tree_model = DecisionTreeClassifier(max_depth=i)

    decition_tree_model.fit(xtrain, ytrain)

   #Traing Data
    train_pred = decition_tree_model.predict(xtrain)
    train_ac = accuracy_score(ytrain,train_pred)
    tr_score.append(train_ac)

    #Test Data
    test_pred = decition_tree_model.predict(xtest)
    test_ac = accuracy_score(ytest,test_pred)
    te_score.append(test_ac)

    print('Depth = %d,train score = %f, test score = %f'%(i,train_ac,test_ac))



```

# Voting (Ensemble Tecnique)

```python
from sklearn.ensemble import VotingClassifier


from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm


b_rf = RandomForestClassifier()
b_ada = AdaBoostClassifier()
b_xg = XGBClassifier()
b_gra = GradientBoostingClassifier()
b_knn = KNeighborsClassifier(n_neighbors=2)
b_lor = LogisticRegression()
b_nb = GaussianNB()
b_dt = DecisionTreeClassifier()
b_sv = svm.SVC(probability=True)

vc = VotingClassifier(estimators=[('cl1',b_rf), ('cl2', b_ada), ('cl3', b_xg), ('cl4', b_gra), ('cl5',b_knn), ('cl6',b_lor), ('cl7',b_nb), ('cl8',b_dt), ('cl9',b_sv)], voting='hard')


vc.fit(xtrain,ytrain)
y_predict_ensem = vc.predict(xtest)

from sklearn import metrics
vc_acc = metrics.accuracy_score(ytest,y_predict_ensem)*100

print("Voting Ensembol Classifier Accuracy: ",'%.2f'%vc_acc)

vc
```

# Performance Study of Machine Learning Algorithms before using (PCA)

```python
from sklearn import metrics

# Build Confusion Metrics
#ConMat = metrics.confusion_matrix(ytest, y_predict_ensem)
ConMat = metrics.confusion_matrix(ytest, y_pre_rf)


#Extract TN, FP, TP from Multidimentional array
# TN, FP, FN, TP = metrics.confusion_metrix(ytest, y_pre_rf).ravel()
TN,FP,FN,TP = metrics.confusion_matrix(ytest,y_pre_rf).ravel()

#Calculate Accuracy 
#Accuracy = metrics.accuracy_score(ytest,y_predict_ensem)*100
Accuracy = metrics.accuracy_score(ytest, y_pre_rf)*100

# All Formula

Prec = TP/(TP+FP)
TPR  = TP/(TP+FN)
FPR  = FP/(TN+FP)
TNR  = TN/(TN+FP)
FNR  = FN/(FN+TP)


#Print the  Result

print("Confusion Matrix: \n",ConMat)
print('True Negative: ', TN)
print('Flase Positive: ',FP)
print('False Negative: ',FN)
print('True Positive:', TP)
print('Accuracy: %.2f'%Accuracy)
print('Precision: %.2f'%Prec)
print('Recall/True Positive Rate: %.2f' %TPR)
print('Flase Positive Rate: %.2f' %FPR)
print('Specificity/True Negative Rate: %.2f' %TNR)
print('Flase Negative Rate: %.2f'%FNR)
```

# ROC curve, AUC
An ROC curve (receiver operating characteristic curve) is a graph showing the performance of a classification model at all classification thresholds. This curve plots two parameters:

True Positive Rate
False Positive Rate
True Positive Rate (TPR) is a synonym for recall and is therefore defined as follows:
TRP = TP/(TP+FN)

False Positive Rate (FPR) is defined as follows:
FRP = FP/(FP/TN) 


```python
from sklearn.metrics import roc_curve,auc
from matplotlib import pyplot as plt

fpr, tpr, treshold = roc_curve (ytest, y_predict_ensem)
auc_sc = auc(fpr, tpr)

plt.figure(figsize=(5,5),dpi=100) #dpi = dots per inch
plt.plot(fpr,tpr,linestyle='-',label = 'model_auc(auc=%0.2f)'%auc_sc)
plt.plot([0,1],[0,1], color='orange',linestyle='--')


plt.xlabel('False Positive Rate-->')
plt.ylabel('True Positive Rate-->')
plt.legend()
plt.show()
```

```python
from sklearn import metrics

# Build Confusion Metrics
#ConMat = metrics.confusion_matrix(ytest, y_predict_ensem)
ConMat = metrics.confusion_matrix(ytest, y_pre_rf)


#Extract TN, FP, TP from Multidimentional array
# TN, FP, FN, TP = metrics.confusion_metrix(ytest, y_pre_rf).ravel()
TN,FP,FN,TP = metrics.confusion_matrix(ytest,y_pre_rf).ravel()

#Calculate Accuracy 
#Accuracy = metrics.accuracy_score(ytest,y_predict_ensem)*100
Accuracy = metrics.accuracy_score(ytest, y_pre_rf)*100

# All Formula

Prec = TP/(TP+FP)
TPR  = TP/(TP+FN)
FPR  = FP/(TN+FP)
TNR  = TN/(TN+FP)
FNR  = FN/(FN+TP)


#Print the  Result

print("Confusion Matrix: \n",ConMat)
print('True Negative: ', TN)
print('Flase Positive: ',FP)
print('False Negative: ',FN)
print('True Positive:', TP)
print('Accuracy: %.2f'%Accuracy)
print('Precision: %.2f'%Prec)
print('Recall/True Positive Rate: %.2f' %TPR)
print('Flase Positive Rate: %.2f' %FPR)
print('Specificity/True Negative Rate: %.2f' %TNR)
print('Flase Negative Rate: %.2f'%FNR)
```
